```python
numbers = [1, 2, 3]
extra = [4, 5]

numbers.append(extra)   # adds the whole list as one element
print("append:", numbers)

numbers = [1, 2, 3]
numbers.extend(extra)   # adds elements individually
print("extend:", numbers)
```

```python
cart = ["apple", "banana", "apple", "mango", "banana", "grapes"]
# 1. Count how many times "apple" appears.
print("Apple count:", cart.count("apple"))
# 2. Remove duplicates using only list methods
unique_cart = []
for item in cart:
    if item not in unique_cart:   # check before adding
        unique_cart.append(item)

# 3. Sort alphabetically
unique_cart.sort()

print("Final cart:", unique_cart)
```

```python
cart = ["milk", "bread", "butter", "eggs"]

cart.extend(["jam", "cheese"])     # add multiple
cart.insert(0, "honey")            # insert at beginning
cart.remove("bread")               # remove bread
cart.pop()                         # remove last item
cart.sort()                        # sort alphabetically

print("Final Cart:", cart)
print("Milk count:", cart.count("milk"))
```

```python
marks = [89, 56, 92, 75, 56, 92, 100, 67]

marks.sort()
print("Sorted:", marks)

second_highest = marks[-2]
print("Second highest:", second_highest)

lowest = marks[0]
while lowest in marks:
    marks.remove(lowest)
print("After removing lowest:", marks)

marks.extend([85, 90])
print("After adding:", marks)

marks.reverse()
print("Reversed:", marks)
```

```python
marks = [45, 67, 89, 23, 90, 56]

# 1. Total marks
total = 0
for m in marks:
    total += m
print("Total Marks:", total)

# 2. Highest and Lowest
highest = marks[0]
lowest = marks[0]
for m in marks:
    if m > highest:
        highest = m
    if m < lowest:
        lowest = m
print("Highest:", highest, "Lowest:", lowest)

# 3. Count >= 50
count = 0
for m in marks:
    if m >= 50:
        count += 1
print("Students scoring >=50:", count)
```

```python
inventory = ["pen", "notebook", "eraser", "pen", "marker", "notebook"]

pen_count = inventory.count("pen")

unique_inventory = []
for item in inventory:
    if item not in unique_inventory:
        unique_inventory.append(item)

unique_inventory.sort()


print("Pen count:", pen_count)
print("Final Inventory:", unique_inventory)

```

```python
numbers = [1, 2, 3]
numbers[3:] = [4, 5, 6]   # inserting at the end
print(numbers)
# Output: [1, 2, 3, 4, 5, 6]

# Insert in the middle
numbers[2:2] = [7, 8]
print(numbers)
# Output: [1, 2, 7, 8, 3, 4, 5, 6]
```

```python
numbers = [1, 2, 3]
for val in [4, 5, 6]:
    numbers.insert(len(numbers), val)   # insert at end
print(numbers)
# Output: [1, 2, 3, 4, 5, 6]
```
