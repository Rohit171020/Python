```python
# Q.9.Check if a number is a palindrome .
# 121-> 121
# 122-> 221 (not a palindrome)

num = int(input("enter a number"))
rev = 0
n = num
while n>0:
    digit = n%10
    rev = rev*10 + digit
    n = n//10
if rev == num:
    print("number is palindrome")
else:
    print('number is not a palindrome')


```

```python
# Check Armstrong number (e.g., 153 = 1³+5³+3³).
# digits = len(str(num))

# 153 = 3

num = int(input("enter the number"))
pow = 0
n = num
digits = len(str(num))
while n>0:
    digit = n%10
    pow = pow + digit**digits
    n = n//10
if pow == num:
    print("Armstrong number")
else:
    print('not armstrong')
```

```python
# Q5. Rock–Paper–Scissors Game
#  Question: Create a Python game where the user 
# plays Rock-Paper-Scissors against the computer.
# * The computer’s choice should be random.
# The game should continue until the user types "quit".

import random
choices = ["rock","paper","scissor"]
while True:
    user = input("Ente the choice (rock,paper,scissor,quit)")
    if user == "quit":
        print("program is ended")
        break
    if user not in choices:
        print("invalid choice")
        continue
    computer = random.choice(choices)
    print("choose the choice",computer)
    if user == computer:
        print("it is tie")
    elif (user== "rock" and computer=="scissor") or\
    (user == "paper" and computer == "rock") or\
    (user== "scissor" and computer == "paper"):
        print("user win")
    else:
        print("Computer win")



```

```python

counter = 0  

def conditional_increment(value):
    global counter
    if value >0:
        counter += value
        print("Counter incremented by", value, ". New counter value:", counter)
    else:
        print("increment value must be positive")

# Step 3: # Call the function with different values to test the behavior
conditional_increment(5)   # Should increment the counter by 5
conditional_increment(-3)  # Should print an error message
conditional_increment(10)  # Should increment the counter by 10

# Printing the final value of the counter after all function calls
print("Final counter value:", counter)


```
