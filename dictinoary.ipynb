```python
s = "mississippi"
freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

max_char = max(freq, key=freq.get)
print("Most frequent:", max_char, "Count:", freq[max_char])
```

```python
for name, subjects in students.items():
    total = 0
    count = 0
    for marks in subjects.values():
        total += marks
        count += 1
    average = total / count
    print(name, "Average Marks:", average)
```

```python
max_marks = 0
topper = ""
for name, subjects in students.items():
    total = sum(subjects.values())
    if total > max_marks:
        max_marks = total
        topper = name

print("Topper:", topper, "with Total Marks:", max_marks)
```

```python
min_marks = None
lowest = ""
for name, subjects in students.items():
    total = sum(subjects.values())
    if min_marks is None or total < min_marks:
        min_marks = total
        lowest = name

print("Lowest Scorer:", lowest, "with Total Marks:", min_marks)
```

```python
above_80_all = []

for name, subjects in students.items():
    all_above_80 = True
    for marks in subjects.values():
        if marks <= 80:
            all_above_80 = False
            break
    if all_above_80:
        above_80_all.append(name)

print("Students scoring above 80 in all subjects:", above_80_all)
```

```python
less_70_any = []
for name, subjects in students.items():
    for marks in subjects.values():
        if marks < 70:
            less_70_any.append(name)
            break

print("Students scoring less than 70 in any subject:", less_70_any)
```
