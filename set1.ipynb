```python
data = [(1,"a"), (2,"b"), (1,"c"), (3,"d"), (2,"e")]

seen = set()
duplicates = set()

# Step 1: Identify duplicates
for x in data:
    if x in seen:
        duplicates.add(x)
    else:
        seen.add(x)

# Step 2: Keep only tuples not in duplicates
result = []
for t in data:
    if t[0] not in duplicates:
        result.append(t)

print(result)   # [(3, 'd')]

```

```python
data = [(1,"a"), (2,"b"), (1,"c"), (3,"d"), (2,"e")]

# Step 1: Count how many times each first element appears
freq = {}
for tup in data:             # take each tuple
    num = tup[0]             # first element of tuple
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

# Step 2: Keep only tuples whose first element occurs once
result = []
for tup in data:
    num = tup[0]
    if freq[num] == 1:
        result.append(tup)

print(result)   #
```

```python
# Question 1: List Processing with Loops.Write a 
# function that takes a list of numbers and returns a 
# new list containing only the even numbers, squared.
# Example: [1, 2, 3, 4, 5, 6] → [4, 16, 36]
def square_number(numbers):
    result = []
    for num in numbers:
        if num%2==0:
            result.append(num**2)
    return result
square_number([1, 2, 3, 4, 5, 6])
```

```python
# Create a function that generates a multiplication 
# table as a list of lists, where each inner list represents a row
# .Example 
# (size = 3)
# [[1, 2, 3], [2, 4, 6], [3, 6, 9]]
def multipication_table(size):
    table = []
    for i in range(1,size+1):
        row = []
        for j in range(1,size+1):
            row.append(i*j)
        table.append(row)
    return table
multipication_table(3)
```

```python
# Given a list of words, create a 
# function that returns a dictionary
# where keys are the first letters and 
# values are sets of words starting with that letter.
# Example: ['apple', 'banana', 'cherry', 'apricot', 'blueberry'] → 
# {'a': {'apple', 'apricot'}, 'b': {'banana', 'blueberry'}, 'c': {'cherry'}}
def group_first_letter(words):
    result = {}
    for word in words:
        first_letter = word[0].lower()
        if first_letter not in result:
            result[first_letter] = set()
        result[first_letter].add(word)
    return result
print(group_first_letter(['apple', 'banana', 'cherry', 'apricot', 'blueberry']))
    
```

```python
class Solution:
    def isPowerOfTwo(self, n: int) -> bool:
        if n <= 0:
            return False
        while n >1:
            if n % 2 != 0:  
                return False
            n //= 2
        return True
```

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        n = len(s)
        max_len = 0

        for i in range(n):
            seen = set()
            for j in range(i, n):
                if s[j] in seen:
                    break   # duplicate found, stop this substring
                seen.add(s[j])
                max_len = max(max_len, j - i + 1)

        return max_len

        
```

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        freq = {}
        n = len(nums)
        for num in nums:
            freq[num] = freq.get(num,0)+1
            if freq[num]>n//2:
                return num
```

```python
class Solution:
    def areOccurrencesEqual(self, s: str) -> bool:
        freq = {}
        for ch in s:
            if ch in freq:
                freq[ch]+=1
            else:
                freq[ch] = 1
        values = list(freq.values())
        for count in values:
            if count!=values[0]:
                return False
        return True
```

```python
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        length = 0
        i = len(s) - 1

        # Step 1: Skip trailing spaces
        while i >= 0 and s[i] == ' ':
            i -= 1

        # Step 2: Count the last word
        while i >= 0 and s[i] != ' ':
            length += 1
            i -= 1

        return length
```

```python
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        left = 0
        right = 0
        res = []
        while left<m and right<n:
            if nums1[left]<nums2[right]:
                res.append(nums1[left])
                left+=1
            else:
                res.append(nums2[right])
                right+=1
        while left<m:
            res.append(nums1[left])
            left+=1
        while right<n:
            res.append(nums2[right])
            right+=1
        for i in range(len(res)):
            nums1[i] = res[i]


```

```python
FUNCTION groupAnagrams(strs):

    CREATE an empty dictionary anagram

    FOR each word in strs:
        sorted_word = sorted letters of the word (as a tuple)
        
        IF sorted_word exists in anagram:
            ADD word to anagram[sorted_word]
        ELSE:
            CREATE new entry in anagram with key = sorted_word
            SET its value as a list containing the word

    CREATE an empty list result

    FOR each key in anagram:
        ADD anagram[key] to result

    RETURN result
```

```python
FUNCTION isPalindrome(s)
    left ← 0
    right ← LENGTH(s) - 1

    WHILE left < right DO
        // skip non-alphanumeric from left
        WHILE left < right AND s[left] is NOT alphanumeric DO
            left ← left + 1
        END WHILE

        // skip non-alphanumeric from right
        WHILE left < right AND s[right] is NOT alphanumeric DO
            right ← right - 1
        END WHILE

        // compare lowercase characters
        IF LOWERCASE(s[left]) ≠ LOWERCASE(s[right]) THEN
            RETURN False
        END IF

        // move pointers
        left ← left + 1
        right ← right - 1
    END WHILE

    RETURN True
END FUNCTION
```

```python
def sum_digits(num):
    total = 0
    while num > 0:
        while total > 9:
            total = total + num % 10
            num = num // 10
            total = total // 10
    return total

print(sum_digits(38))
print(sum_digits(123))
```
