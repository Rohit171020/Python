```python
# Q7. Write a Python program to find all duplicate elements in a tuple.
t = (4, 6, 2, 6, 7, 4, 9, 2, 6)
duplicates = []
for x in t:
    if t.count(x)>1 and x not in duplicates:
        duplicates.append(x)
print(tuple(duplicates))
```

```python
# Q8.find the intersection of 2 tuples without using sets:
t1 = (1, 2, 3, 4, 5, 6)
t2 = (4, 5, 6, 7, 8, 9)
intesection = []
for x in t1:
    if x in t2 and x not in intesection:
        intesection.append(x)
print(intesection)
```

```python
# Q9.Find element(s) that appear only once in
# a tuple. t = (4, 5, 6, 4, 7, 8, 6, 9)
t = (4, 5, 6, 4, 7, 8, 6, 9)
unique = []
for x in t:
    if t.count(x)==1:
        unique.append(x)
print(tuple(unique))
```

```python
# Q10 Find all unique pairs of numbers in a tuple whose sum is 10.
t = (1, 2, 3, 7, 8, 9, 5, 5)
pairs = []
for i in range(len(t)):
    for j in range(i+1,len(t)):
        if t[i]+ t[j] == 10:
            pairs.append((t[i],t[j]))
print(pairs)
```
