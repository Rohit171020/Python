```python
marks = [89, 56, 92, 75, 56, 92, 100, 67]

# Step 1: Sort ascending
marks.sort()
print("Ascending order:", marks)

# Step 2: Find second highest
unique_marks = sorted(set(marks))   # remove duplicates
second_highest = unique_marks[-2]
print("Second highest:", second_highest)

# Step 3: Remove all occurrences of lowest mark
lowest = min(marks)
while lowest in marks:
    marks.remove(lowest)
print("After removing lowest:", marks)

# Step 4: Add [85, 90]
marks.extend([85, 90])
print("After adding new marks:", marks)

# Step 5: Reverse for highest first
marks.sort(reverse=True)  # or marks.reverse() after sorting
print("Highest first:", marks)

```

```python

nums = [12, 7, 9, 21, 14, 18, 3, 28]

# 1. Print only even numbers
print("Even numbers:", [n for n in nums if n % 2 == 0])

# 2. Find second largest manually (no set, no sorted)
largest = second_largest = float('-inf')
for n in nums:
    if n > largest:
        second_largest = largest
        largest = n
    elif n > second_largest and n != largest:
        second_largest = n

print("Second largest:", second_largest)

# 3. Reverse the list
nums.reverse()
print("Reversed list:", nums)

```

```python
# Take user input
user_input = input("")

# Split input into list
raw_list = user_input.split(",")

# Create an empty list to store cleaned destinations
destinations_list = []
# Loop through each item in raw_list
for place in raw_list:
    cleaned_place = place.strip()   # remove extra spaces
    destinations_list.append(cleaned_place)
# Convert list into tuple
destinations_tuple = tuple(destinations_list)
# Print final tuple
print("Tuple of favorite travel destinations:", destinations_tuple)
```

```python
def create_tuple():
    # Take user input
    user_input = input("")
    
    # Split input into raw list
    raw_list = user_input.split(",")
    
    # Clean spaces using a loop
    cleaned_list = []
    for item in raw_list:
        cleaned_list.append(item.strip())
    
    # Convert list into tuple
    return tuple(cleaned_list)


# Create first tuple
print("")
tuple1 = create_tuple()

# Create second tuple
print("")
tuple2 = create_tuple()

# Concatenate inside a nested tuple
nested_tuple = (tuple1, tuple2)

# Print final nested tuple
print("Nested Tuple:", nested_tuple)
```
