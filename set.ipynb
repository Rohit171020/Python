```python
def multiplication_table(size):
    table = []
    for i in range(1, size + 1):
        row = []
        for j in range(1, size + 1):
            row.append(i * j)
        table.append(row)
    return table

# Test
print(multiplication_table(3))
```

```python
def compare_lists(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    
    common = set1.intersection(set2)
    only_in_first = set1.difference(set2)
    only_in_second = set2.difference(set1)
    
    return (common, only_in_first, only_in_second)

# Test
result = compare_lists([1, 2, 3, 4], [3, 4, 5, 6])
print(result)  # ({3, 4}, {1, 2}, {5, 6})
```

```python
def swap_dict_keys_values(original_dict):
    """
    Swaps keys and values in a dictionary, handling duplicate values.
    
    Args:
        original_dict (dict): Dictionary to swap
    
    Returns:
        dict: New dictionary with swapped keys and values
    """
    swapped = {}
    
    for key, value in original_dict.items():
        if value in swapped:
            # Value already exists as key, add to list
            swapped[value].append(key)
        else:
            # New value, create list with current key
            swapped[value] = [key]
    
    return swapped

# Test cases
print("Test 1:", swap_dict_keys_values({'name': 'John', 'age': 25}))
print("Test 2:", swap_dict_keys_values({'Alice': 'A', 'Bob': 'B', 'Charlie': 'A'}))
```
